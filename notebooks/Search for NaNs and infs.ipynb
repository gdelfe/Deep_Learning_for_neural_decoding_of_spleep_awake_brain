```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
# loop through data, see if they are nan and inf

movement_dir = 'data/moving/'
```

```python
bad_recs = []
for f in os.listdir(movement_dir):
    sample = np.load(movement_dir + f)

    if np.isnan(sample.sum()):
        if f.split('_')[:2] not in bad_recs:
            bad_recs.append(f.split('_')[:2])
        #print("Noah you messed up")
        print(f)
        #break
```
