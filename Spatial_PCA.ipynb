# ################PCA###############

```python
def spatial_pca (spectrogram, n_comp):
    #Reshape movement to an electrode * frequency array
    spec = []
    for e in range(spectrogram.shape[0]):
        temp = spectrogram[e].flatten()
        spec.append(temp)
    spec = np.array(spec)
    spec = np.array(np.hsplit(spec,10))#time*electrode*freq 
    #Perform spatial PCA to reduce number of electrodes
    pca = PCA(n_components = n_comp)
    spec_red=[]
    for t in range(spec.shape[0]):
        temp = spec[t].transpose()
        spec_temp= pca.fit_transform(temp)
        spec_temp=spec_temp.transpose()
        spec_red.append(spec_temp)
    spec_red = np.array(spec_red)
    #Reshape movement to an time * frequency array
    spec_PCA= []
    for t in range(spec_red.shape[0]):
        temp = spec_red[t].flatten()
        spec_PCA.append(temp)
    spec_PCA = np.array(spec_PCA)
    spec_PCA = np.array(np.hsplit(spec_PCA,20))
    return spec_PCA
```

```python
for m in move_file_window:
    move = np.load(os.path.join(move_file_path, m))
    spec_PCA = spatial_pca(move,20)
    ##Save files
    filename = m.split('.')
    filename_PCA = filename[0]+'_PCA.'+filename[1] 
#     file_save = '/mnt/pesaranlab/People/Capstone_students/Spectrogram_PCA_data/'+filename_PCA
#     np.save(file_save, spec_PCA)
    
```
