## convLSTM

```python
device = torch.device('cuda')

tol = 0.01
learning_rate = 0.001
num_epochs = 51
alpha = 0
model_type = 'LR'
loss_type = 'bce'
reg_type = 'none'

```

```python
class convLSTM(nn.module):
    def __init__(self, input_dim = 62*100*10, output_dim = 1, hidden_dim, kernel_size, bias = None):
        super(convLSTM, self).__init__()

        self.input_dim = input_dim
        self.hidden_dim = hidden_dim # Number of hidden layers

        self.kernel_size = kernel_size
        self.padding = kernel_size[0] // 2, kernel_size[1] // 2
        self.bias = bias
        
        #-----initialise weights and biases-----#
        self.params = {}
        std = (1.0/np.sqrt(self.input_dim + self.hidden_dim)) # Xavier initialisation
        
        # forget gate
        self.params["Wf"] = np.random.randn(self.hidden_dim, self.hidden_dim + self.input_dim) * std
        self.params["bf"] = np.ones((self.hidden_dim,1))

        # input gate
        self.params["Wi"] = np.random.randn(self.hidden_dim, self.hidden_dim + self.input_dim) * std
        self.params["bi"] = np.zeros((self.hidden_dim,1))

        # cell gate
        self.params["Wc"] = np.random.randn(self.hidden_dim, self.hidden_dim + self.input_dim) * std
        self.params["bc"] = np.zeros((self.hidden_dim,1))

        # output gate
        self.params["Wo"] = np.random.randn(self.hidden_dim, self.hidden_dim + self.input_dim) * std
        self.params["bo"] = np.zeros((self.hidden_dim ,1))

        # output
        self.params["Wv"] = np.random.randn(self.output_dim, self.hidden_dim) * \
                                          (1.0/np.sqrt(self.input_dim))
        self.params["bv"] = np.zeros((self.output_dim ,1))
        
        def forward_step(self, x, h_prev, c_prev):
            z = np.row_stack((h_prev, x))

            f = self.sigmoid(np.convolve(self.params["Wf"], z) + self.params["bf"])
            i = self.sigmoid(np.convolve(self.params["Wi"], z) + self.params["bi"])
            c_bar = np.tanh(np.convolve(self.params["Wc"], z) + self.params["bc"])

            c = f * c_prev + i * c_bar
            o = self.sigmoid(np.convolve(self.params["Wo"], z) + self.params["bo"])
            h = o * np.tanh(c)

            v = np.dot(self.params["Wv"], h) + self.params["bv"]
            y_hat = self.softmax(v)
            return y_hat, v, h, o, c, c_bar, i, f, z
```
